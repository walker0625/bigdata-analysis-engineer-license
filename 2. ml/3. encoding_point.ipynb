# 인코딩

```python
import pandas as pd

train = pd.read_csv('https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p2/ch2/train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p2/ch2/test.csv')
```

```python
target = train.pop('income')
target
```

# 원핫인코딩(유니크한 값들을 별도의 컬럼으로 분리 - T/F)

```python
train_oh = pd.get_dummies(train)
test_oh = pd.get_dummies(test)

# y값 외에도 unique 값이 다른 상황
print(train.shape, train_oh.shape, test.shape, test_oh.shape)
```

```python
total_data = pd.concat([train, test], axis=0)
total_data.shape

data_oh = pd.get_dummies(total_data)

train_oh = data_oh.iloc[:len(train)]
test_oh = data_oh.iloc[len(train):]

print(train.shape, train_oh.shape, test.shape, test_oh.shape)
```
