# 스케일링(수치형 데이터간의 범위 비대칭 보정)

```python
import pandas as pd

train = pd.read_csv('https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p2/ch2/train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p2/ch2/test.csv')

train.info()
```

```python
from sklearn.preprocessing import MinMaxScaler # 0~1 사이의 값으로 변환

cols =['age', 'education.num']

scaler = MinMaxScaler()

train[cols] = scaler.fit_transform(train[cols])
test[cols] = scaler.transform(test[cols])

train.head()
```

```python
from sklearn.preprocessing import StandardScaler # 평균이 0 표준편차 1인 정규분포

scaler = StandardScaler()

train[cols] = scaler.fit_transform(train[cols])
test[cols] = scaler.transform(test[cols])

train.head()
```

```python
from sklearn.preprocessing import RobustScaler # 중앙값과 사분위값 활용, 이상치가 많은 경우 활용

scaler = RobustScaler()

train[cols] = scaler.fit_transform(train[cols])
test[cols] = scaler.transform(test[cols])

train.head()
```

```python
total = pd.concat([train, test], axis=0)
total.shape
```

```python
total['income'] = total['income'].replace('>50K', 1).replace('<=50K', 0)
total['income']
```
