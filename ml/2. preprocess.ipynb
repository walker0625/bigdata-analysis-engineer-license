# 결측치 처리(기출에 나온적은 없으나 있으면 삭제하고 진행)

```python
import pandas as pd

train = pd.read_csv('https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p2/ch2/train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p2/ch2/test.csv')
```

```python
# 삭제

# df = train.dropna() # 결측이 있는 행을 삭제 - test 데이터는 행 삭제하면 0점됨 주의
# df = train.dropna(axis=1) # 결측이 있는 열을 삭제 - 삭제는 가능한데 할꺼면 train/test 둘다 삭제
```

```python
# 채우기
# 1. 문제의 평가 지표의 결과에 따라서 뭘로 채울지는 결정 - 잘나온 결과에 따라서 결정
# 2. 전처리를 할꺼면 test도 똑같이 처리해야 함
# 3. train에서 생성한 값으로 train/test 둘 다 채워도 무방

# 범주형
most = train['workclass'].mode()[0] # 최빈값

train['workclass'] = train['workclass'].fillna(most)

most
```

```python
# 수치형
mean = int(train['age'].mean())
print(mean)

train['age'] = train['age'].fillna(mean)

median = train['hours.per.week'].median()
train['hours.per.week'] = train['hours.per.week'].fillna(median)
print(median)
```
