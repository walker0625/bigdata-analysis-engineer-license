```python
import pandas as pd

df = pd.read_csv('data6-1-1.csv')
df.head(1)
```

```python
df.info()
```

```python
df['출동시간'] = pd.to_datetime(df['출동시간'])
df['도착시간'] = pd.to_datetime(df['도착시간'])
df.head()
```

```python
df['차이'] = df['도착시간'] - df['출동시간']
df
```

```python
df['차이'] = df['차이'].dt.total_seconds() / 60
```

```python
df.groupby('소방서').mean()
```

```python
result = df.groupby('소방서').mean().sort_values('차이', ascending=False)
result
```

```python
print(round(result.iloc[0]['차이']))
```
