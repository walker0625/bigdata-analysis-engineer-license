# 문제3

```python
df = pd.read_csv('members.csv')
df.head()
```

```python
print(df.shape)
df = df.dropna(subset = ['views'])
print(df.shape)
```

```python
# 1) replace

import numpy as np

print(df.head(8))
df['f3'] = df['f3'].replace(np.nan, 0).replace('silver', 1).replace('gold', 2).replace('vip', 3)
print(df.head(8))
```

```python
print(df['f3'].sum())
```

```python
# 2) map
df = pd.read_csv('members.csv')

dict_list = {np.nan: 0, 'silver': 1, 'gold': 2, 'vip': 3}
df['f3'] = df['f3'].map(dict_list)
print(df['f3'].sum())
```

```python
# 3) condition

df = pd.read_csv('members.csv')

r1 = sum(df['f3'] == 'silver') * 1
r2 = sum(df['f3'] == 'gold') * 2
r3 = sum(df['f3'] == 'vip') * 3

print(r1 + r2 + r3)
```

# 문제5

```python
df = pd.read_csv('members.csv')
df.head()
```

```python
q3 = df['views'].quantile(.75)
q1 = df['views'].quantile(.25)
print(q3)
print(q1)

IQR = q3 - q1
print(IQR)

low_line = q1 - 1.5 * IQR
high_line = q3 + 1.5 * IQR

print(sum(df['views'] < low_line))
print(sum(df['views'] > high_line))

print(sum(df['views'] < low_line) + sum(df['views'] > high_line))
```

# 문제 7

```python
import pandas as pd
import random
random.seed(2022)
df = pd.DataFrame()
for i in range(0, 5):
    list_box = []
    for k in range(0, 200):
        ran_num = random.randint(1,200)
        list_box.append(ran_num)
    df[i+2000] = list_box
df = df.T
df.to_csv("data.csv", index=True)
```
