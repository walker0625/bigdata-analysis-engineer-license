# Basic

```python
import pandas as pd
```

```python
color = pd.Series(['빨강', '파랑', '노랑'])
color
```

```python
price = pd.Series([1000, 2000, 3000])
price
```

```python
tone = pd.Series([10, 20, 30])
tone
```

```python
df = pd.DataFrame({
    '색상' : color,
    '가격' : price,
    '톤' : tone
})

df
```

```python
# Series 형태로 선택하는 경우 []
type(df['색상'])
```

```python
# DataFrame 형태로 선택하는 경우
type(df)
```

```python
# DataFrame 형태로 선택하는 경우 [[]]
type(df[['색상']])
```

```python
# DataFrame 형태로 선택하는 경우 [[]]
type(df[['색상', '가격']])
```

```python
df.to_csv('test.csv', index=False) # index = False 하지 않으면 index를 별도의 컬럼으로 또 저장함(Unnamed)
```
