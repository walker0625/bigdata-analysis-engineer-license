```python
# client num은 의미가 없으므로 삭제(단순 index 값임)
train = train.drop('CLIENTNUM', axis=1)
test_id = test.pop('CLIENTNUM') # 최종 제출할 파일에 들어가야 하므로 빼둠

target = train.pop('Attrition_Flag')
target
```

```python
train = pd.get_dummies(train)
test = pd.get_dummies(test)
train
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    train,
    target,
    test_size=0.2,
    random_state=0
)

X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)
```

```python
pred = rf.predict(X_val)
pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

print(accuracy_score(y_val, pred))
print(precision_score(y_val, pred))
print(recall_score(y_val, pred))
print(f1_score(y_val, pred))
```

```python
pred_proba = rf.predict_proba(X_val)
print(pred_proba)

# [[0.97 0.03]
#  [0.96 0.04]
#  [0.9  0.1 ]
#  ...
#  [0.59 0.41]
#  [0.95 0.05]
#  [0.87 0.13]]
print(roc_auc_score(y_val, pred_proba[:, 1])) # 전체에서 2번째 선택 0.03, 0.04
```
