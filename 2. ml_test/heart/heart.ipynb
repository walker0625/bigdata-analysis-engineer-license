```python
import pandas as pd

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train.nunique()
```

```python
train = train.drop('id', axis=1)
train
```

```python
test_id = test.pop('id')
test_id
```

```python
from sklearn.model_selection import train_test_split

target = train.pop('output')

X_train, X_val, y_train, y_val = train_test_split(
    train,
    target,
    test_size=0.2,
    random_state=0
)

X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

# 하이퍼 파라미터 튜닝 / max_depth=3~12, n_estimators=200~1000
rf = RandomForestClassifier(random_state=0, max_depth=5, n_estimators=200)
rf.fit(X_train, y_train)
```

```python
pred = rf.predict(X_val)
pred

pred_proba = rf.predict_proba(X_val)
pred_proba
```

```python
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score

print(roc_auc_score(y_val, pred_proba[:, 1]))
print(f1_score(y_val, pred))
print(accuracy_score(y_val, pred))

# rf
# 0.9297658862876255
# 0.8679245283018868
# 0.8571428571428571

# rf max_depth=3 / n_estimators=200
# 0.9565217391304348
# 0.8888888888888888
# 0.8775510204081632
```
